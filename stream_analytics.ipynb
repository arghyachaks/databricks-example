```
lines = spark.readStream.format('socket').option('host', '54.213.33.240').option('port', 9002).load()
edits = lines.select(json_tuple("value", "channel","timestamp","isRobot","isAnonymous")).selectExpr("c0 as channel","cast(c1 as timestamp) as time","c2 as page").createOrReplaceTempView("edits")
editCounts = spark.sql("""SELECT count(*) channel, date_format(window(time, '10 seconds').start,'HH:mm:ss') as time FROM edits GROUP BY channel, window(time, '10 seconds') ORDER BY time""")
```
